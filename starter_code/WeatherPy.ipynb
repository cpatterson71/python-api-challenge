### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# create the lists in the DataFrame
cities_df['City_Name'] = ""
cities_df['Humidity'] = ""
cities_df['Max_Temp'] = ""
cities_df['Wind_Speed'] = ""
cities_df['Cloudiness'] = ""
cities_df['Date'] = ""

cities_df.head()
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
weather_api_key = 'c69238b9553f803c3dbc5f5427b2aa7a'
count = 0 

for index, row in cities_df.iterrows():
    Lat = str(row['Lat'])
    #
    Lng = str(row['Lng'])
    
    city = row['City']
    
        
    query_url = url + "lat=" + Lat +"&lon=" + Lng + '&units=Imperial' + '&appid=' + weather_api_key  
    weather_response = requests.get(query_url)
    weather_response = weather_response.json()
    
    #pprint(weather_response)
    
    try:
        cities_df.loc[index, 'City_Name'] = weather_response['name']
        cities_df.loc[index, 'Humidity'] = weather_response['main']['humidity']
        cities_df.loc[index, 'Max_Temp'] = weather_response['main']['temp_max']
        cities_df.loc[index, 'Wind_Speed'] = weather_response['wind']['speed']
        cities_df.loc[index, 'Cloudiness'] = weather_response['clouds']['all']
        cities_df.loc[index, 'Date'] = weather_response['dt']
        count += 1
        print(f'The reocrd count is {count} City: {city}')
        
    except (KeyError, IndexError):
        print('Record not found')
        pass
    
len(cities_df)
```

## Linear Regression

```python
def plot_linear_regression(x_values, y_values, title, text_coordinates):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq,text_coordinates,fontsize=14,color="red")
    plt.xlabel("Latitude")
    plt.ylabel(title)
    print(f"The r-squared is: {rvalue}")
    plt.show()
```

```python
northlat_df = weather_data.loc[weather_data["Lat"] >= 0,:]
southlat_df = weather_data.loc[weather_data["Lat"] < 0,:]

northlat_df.dtypes

```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = northlat_df["Lat"].astype(int)
y_values = northlat_df["Max_Temp"].astype(int)
plot_linear_regression(x_values,y_values,'Max_Temp',(6,30))

plt.savefig('../output_data/northernregresstemp.jpg')
plt.show()
```
